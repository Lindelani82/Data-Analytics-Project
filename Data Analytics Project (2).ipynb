# Step 2: Creating new Columns

Let us create column which will be called CostPerTransaction. This column multipy column of CostPerItem with NumberOfItemsPurchased to result Cost Per Transaction

```python
df["CostPerTransaction"]=df["CostPerItem"]*df["NumberOfItemsPurchased"]
```

```python
#we using the same Logic to Calculate Sales Per Transaction
```

```python
df["SalesPerTransaction"]=df["SellingPricePerItem"]*df["NumberOfItemsPurchased"]
```

```python
#Again we using the same Logic to calculate ProfitPerTransaction but now we minising Sales Per Transaction with Cost PerTransaction
```

```python
df["ProfitPerTransaction"]=df["SalesPerTransaction"]-df["CostPerTransaction"]
```

```python
#let us view the data and see new created columns
df.head(2)
```

```python
# Let round off values in ProfitPerTransaction to the nearest whole number
```

```python
df["ProfitPerTransaction"]=df["ProfitPerTransaction"].round()
```

```python
df.head(3)
```

```python
#The dates of this dataset was splited in different columns let us combine them and form one column e.g 20 May 2018
```

```python
df["Dates"]=df["Day"].astype(str)+"-"+df["Month"].astype(str)+"-"+df["Year"].astype(str)
```

```python
#so we needed to convert column consist of intengers and make them as string so that they they able to combine
```

```python
df.head(3)
```

```python
#The column of ItemPerTransaction it in Capital Letters let us change them to lower cases
```

```python
df["ItemDescription"]=df["ItemDescription"].str.lower()
```

```python
df.head(3)
```

```python
#Now we want split the column of ClientKeyWorkds into 3 columns, we then group them into new columns which will be ClientAge,ClientType and TheLenthOfContract
```

```python
df[["ClientAge","ClientType","TheLengthOfContract"]]=df["ClientKeywords"].str.split(",",expand=True)
```

```python
#Let view the data
df.head(4)
```

```python
#Let us drop the column we splitted
df.drop("ClientKeywords",axis=1,inplace=True)
```

```python
df.head(3)
```

```python
#let us clean the data on those new created columns
```

```python
df["ClientAge"]=df["ClientAge"].str.replace("[","")
df["ClientAge"]=df["ClientAge"].str.replace("'","")
df["ClientType"]=df["ClientType"].str.replace("'","")
df["TheLengthOfContract"]=df["TheLengthOfContract"].str.replace("]","")
df["TheLengthOfContract"]=df["TheLengthOfContract"].str.replace("'","")
```
